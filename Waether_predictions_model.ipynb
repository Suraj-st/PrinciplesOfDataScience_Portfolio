```python
def rain_category(target):
    if target <= 0:
        return "no-rain"
    elif 0 < target <= 50:
        return "light-rain"
    elif 50 < target <= 100:
        return "moderate-rain"
    else:
        return "heavy-rain"
data["rain_category"] = data["target"].apply(rain_category)
```
